```python
import numpy as np
import pandas as pd
import json
from geopy.geocoders import Nominatim
import requests
from pandas.io.json import json_normalize
import matplotlib.cm as cm
import matplotlib.colors as colors
import bs4
```

Scrape data from Wikipedia

```python
web_url=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup=BeautifulSoup(web_url,'html.parser')
```


The tabular data is availabe in table and belongs to class="wikitable sortable"So let's extract only table

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
headers="Postcode,Borough,Neighbourhood"
```

```python
table1=""
for tr in My_table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    table1=table1+row1[1:]
```

```python
print(table1)
```

Writing the data in a .csv file for further use

```python
file=open("toronto.csv","wb")
file.write(bytes(table1,encoding="ascii",errors="ignore"))
```

```python
import pandas as pd
df = pd.read_csv('toronto.csv',error_bad_lines=False)
df.columns=["Postalcode","Borough","Neighbourhood"]
```
